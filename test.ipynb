```python
import matplotlib.pyplot as plt

# Wähle einen Index aus dem Dataset
index = 5

# Hole ein Beispiel aus dem Dataset
example = DefineDataset.dataset[index]

# Konvertiere das Bild-Tensor in ein numpy array für die Anzeige
image = example['image'].permute(1, 2, 0).numpy()

# Konvertiere die Masken-Tensoren in numpy arrays für die Anzeige und entferne die zusätzliche Dimension
masks = example['masks'].squeeze().numpy()

# Plotte das Bild und die dazugehörigen Masken
plt.figure(figsize=(15, 10))

# Plotte das Bild
plt.subplot(2, 4, 1)
plt.imshow(image)
plt.title('Bild')

# Plotte die Masken
for i in range(6):
    plt.subplot(2, 4, i + 2)
    plt.imshow(masks[i], cmap='gray')
    plt.title(f'Maske {i+1}')

plt.tight_layout()
plt.show()
```
